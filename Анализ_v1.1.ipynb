```python
# для анализа предлагается взять разбивку по ТН ВЭД кодам до 2х знаков
ved_list = df['Товар (2 знака)'].unique()
print(f'Количество групп товаров для анализа: {len(ved_list)}')
```

```python
df_for_work = df.copy()
df_for_work.drop(['Регион РЭЦ', 'Регионы (РЭЦ) и Союзы', 'ТН ВЭД 10', 'ТНВЭД 6',
       'ТНВЭД 4', 'ТНВЭД 2', 'Расшифровка ТН10', 'Расшифровка ТН6',
       'Расшифровка ТН4', 'Расшифровка ТН2', 'Товар и код ТН ВЭД (10 знаков)',
       'Товар и код ТН ВЭД (6 знаков)', 'Товар и код ТН ВЭД (4 знака)','Нацпроект 2021 v1', 'Отрасль 2021 v1'], inplace=True, axis=1)
df_for_work.shape
```

```python
df_for_work.groupby(['Страна', 'Товар (2 знака)']).sum().round(2)
```

```python
df_for_work.drop_duplicates(inplace=True)
```

```python
# создаем DF, где каждой группе товаров (в соответствии с ТНВЭД до 2 знаков) будет соответствовать период и топ-5 импортирующих стран
best_exp = pd.DataFrame(columns=['tn_ved', 
                                 'period', 
                                 '1_country', 
                                 '1_country_sum', 
                                 '1_country_weight', 
                                 '2_country', 
                                 '2_country_sum', 
                                 '2_country_weight', 
                                 '3_country', 
                                 '3_country_sum', 
                                 '3_country_weight', 
                                 '4_country', 
                                 '4_country_sum', 
                                 '4_country_weight', 
                                 '5_country', 
                                 '5_country_sum', 
                                 '5_country_weight']
)
```

```python
period_list = [' Москва 2015 $', ' Москва 2016 $', ' Москва 2017 $',
       ' Москва 2018 $', ' Москва 2019 $', ' Москва 2020 $', ' Москва 2021 $']

for ved in ved_list:
  for period in period_list:
    new_representation = df_for_work[['Страна', 'Товар (2 знака)', period]][df_for_work['Товар (2 знака)']==ved].groupby('Страна').sum().round(2).sort_values(by=period, ascending=False)[:5]
    total_sum = df_for_work[['Страна', 'Товар (2 знака)',period]][df_for_work['Товар (2 знака)']==ved].groupby('Страна').sum().sum()[period].round(2)
 
    best_exp.loc[len(best_exp.index )] = [ved, period.split(' ')[2], 
                                          new_representation.index[0], 
                                          new_representation[period][0], 
                                          round(new_representation[period][0]/total_sum, 2), 
                                          new_representation.index[1], 
                                          new_representation[period][1], 
                                          round(new_representation[period][1]/total_sum,2), 
                                          new_representation.index[2], 
                                          new_representation[period][2], 
                                          round(new_representation[period][2]/total_sum,2), 
                                          new_representation.index[3], 
                                          new_representation[period][3], 
                                          round(new_representation[period][3]/total_sum,2), 
                                          new_representation.index[4], 
                                          new_representation[period][4], 
                                          round(new_representation[period][4]/total_sum,2)
                                          ]
```

```python
for i in range (len(last_year.index)):
  exp_sum = last_year['1_country_sum'].iloc[i] + last_year['2_country_sum'].iloc[i] + last_year['3_country_sum'].iloc[i] + last_year['4_country_sum'].iloc[i] + last_year['5_country_sum'].iloc[i]
  sum_frendly = last_year['1_country_sum'].iloc[i]*last_year['friendly_1'].iloc[i] + last_year['2_country_sum'].iloc[i]*last_year['friendly_2'].iloc[i] + last_year['3_country_sum'].iloc[i]*last_year['friendly_3'].iloc[i] + last_year['4_country_sum'].iloc[i]*last_year['friendly_4'].iloc[i] + last_year['5_country_sum'].iloc[i]*last_year['friendly_5'].iloc[i]
  last_year['minus_exp_sum'].iloc[i] = exp_sum - sum_frendly
```

```python
for i in range (len(last_year.index)):
  exp_weight = last_year['1_country_weight'].iloc[i] + last_year['2_country_weight'].iloc[i] + last_year['3_country_weight'].iloc[i] + last_year['4_country_weight'].iloc[i] + last_year['5_country_weight'].iloc[i]
  last_year['exp_weight'].iloc[i] = exp_weight
```

```python
for i in range (len(last_year.index)):
  exp_weight = last_year['1_country_weight'].iloc[i] + last_year['2_country_weight'].iloc[i] + last_year['3_country_weight'].iloc[i] + last_year['4_country_weight'].iloc[i] + last_year['5_country_weight'].iloc[i]
  weight_frendly = last_year['1_country_weight'].iloc[i]*last_year['friendly_1'].iloc[i] + last_year['2_country_weight'].iloc[i]*last_year['friendly_2'].iloc[i] + last_year['3_country_weight'].iloc[i]*last_year['friendly_3'].iloc[i] + last_year['4_country_weight'].iloc[i]*last_year['friendly_4'].iloc[i] + last_year['5_country_weight'].iloc[i]*last_year['friendly_5'].iloc[i]
  last_year['minus_exp_weight'].iloc[i] = exp_weight - weight_frendly
```
