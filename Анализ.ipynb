```python
df = pd.read_csv('fts_data.csv', encoding='utf-8', engine='python', decimal=',')
df.head()
```

```python
df.info()
```

```python
df[' Москва 2021 $'] = df[' Москва 2021 $'].apply(lambda x: str(x).replace('\xa0','').replace(',','.').replace('-','0'))
df[' Москва 2021 $'] = df[' Москва 2021 $'].astype(float)
```

```python
df.fillna(0, inplace=True)
```

```python
df.head(10)
```

```python
df['Страна'] = df['Страна'].apply(lambda x: str(x).replace('(','').replace(')','')) #есть, например, FM-МИКРОНЕЗИЯ, ФЕДЕРАТИВНЫЕ ШТАТЫ и FM-МИКРОНЕЗИЯ, (ФЕДЕРАТИВНЫЕ ШТАТЫ)
```

```python
print(f"Товар и код ТН ВЭД (10 знаков) уникальные значения: {len(df['Товар и код ТН ВЭД (10 знаков)'].unique())}")
print(f"Товар и код ТН ВЭД (6 знаков) уникальные значения: {len(df['Товар и код ТН ВЭД (6 знаков)'].unique())}")
print(f"Товар и код ТН ВЭД (4 знака) уникальные значения: {len(df['Товар и код ТН ВЭД (4 знака)'].unique())}")
print(f"Товар (2 знака) уникальные значения: {len(df['Товар (2 знака)'].unique())}")
```

```python
ved_list = df['Товар (2 знака)'].unique() # предлагаю взять ТН 2, так как в данном случае упрощается анализ
```

```python
df_for_work = df
df_for_work.drop(['Регион РЭЦ', 'Регионы (РЭЦ) и Союзы', 'ТН ВЭД 10', 'ТНВЭД 6',
       'ТНВЭД 4', 'ТНВЭД 2', 'Расшифровка ТН10', 'Расшифровка ТН6',
       'Расшифровка ТН4', 'Расшифровка ТН2', 'Товар и код ТН ВЭД (10 знаков)',
       'Товар и код ТН ВЭД (6 знаков)', 'Товар и код ТН ВЭД (4 знака)','Нацпроект 2021 v1', 'Отрасль 2021 v1'], inplace=True, axis=1)
df_for_work.shape
```

```python
df_for_work.groupby(['Страна', 'Товар (2 знака)']).sum().round(2)
```

```python
df_for_work.drop_duplicates(inplace=True)
```

```python
best_exp = pd.DataFrame(columns=['tn_ved', 'perod', '1_country', '1_country_sum', '1_country_weight', '2_country', '2_country_sum', '2_country_weight', '3_country', '3_country_sum', '3_country_weight', '4_country', '4_country_sum', '4_country_weight',  '5_country', '5_country_sum', '5_country_weight'])
```

```python
period_list = [' Москва 2015 $', ' Москва 2016 $', ' Москва 2017 $',
       ' Москва 2018 $', ' Москва 2019 $', ' Москва 2020 $', ' Москва 2021 $']

for ved in ved_list:
  for period in period_list:
    new_representation = df_for_work[['Страна', 'Товар (2 знака)',period]][df_for_work['Товар (2 знака)']==ved].groupby('Страна').sum().round(2).sort_values(by=period,ascending=False)[:5]
    total_sum = df_for_work[['Страна', 'Товар (2 знака)',period]][df_for_work['Товар (2 знака)']==ved].groupby('Страна').sum().sum()[period].round(2)
 
    best_exp.loc[ len(best_exp.index )] = [ved, period.split(' ')[2], new_representation.index[0], new_representation[period][0], round(new_representation[period][0]/total_sum, 2), new_representation.index[1], new_representation[period][1], round(new_representation[period][1]/total_sum,2), new_representation.index[2], new_representation[period][2], round(new_representation[period][2]/total_sum,2), new_representation.index[3], new_representation[period][3], round(new_representation[period][3]/total_sum,2), new_representation.index[4], new_representation[period][4], round(new_representation[period][4]/total_sum,2)]
```
